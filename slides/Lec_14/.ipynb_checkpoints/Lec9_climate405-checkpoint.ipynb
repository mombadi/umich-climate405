## <center> Gradient Descent </center>

It is a first order iterative algorithm for mathematical optimization of a function. It is based on taking steps in the opposite direction of the gradient.

**Gradient:** For a differentiable function _f_, the gradient is the derivative of the function ($f'$, $\nabla$), which gives the direction in which the function increases the most and the rate of increase. 

### Example 1: 2D

Let's look at an example 

<span style="font-size: 24px;"> $y = f(x) = x^4 - 3*x^3 + 2 $ <span>

```python
# Define the function and its derivative
def f(x):
    return x**4 - 3*x**3 + 2

def f_prime(x):
    return 4*x**3 - 9*x**2
```

Let's plot the function:

```python
# Generate the x values and the corresponding y values for the function plot
x_vals = np.linspace(-2, 3, 500)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('y')
```

Now, we're going to code a gradient descent algorithm:

```python
# Gradient descent function
def gradient_descent(start_x, learning_rate, num_steps):
    x = start_x
    path = [x]
    gradient = []
    for _ in range(num_steps):
        gradient.append(f_prime(x))
        x -= learning_rate * f_prime(x)
        path.append(x)
    return np.array(path), np.array(gradient)
```

```python
# Set up the initial point, learning rate, and number of steps
start_x = 2.5
learning_rate = 0.01
num_steps = 100

# Perform gradient descent
path, gradient = gradient_descent(start_x, learning_rate, num_steps)

# Generate y values for the path of gradient descent
path_y_vals = f(path)

# Create the plot using Plotly
fig = go.Figure()

# Plot the function
fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', name='f(x)', line=dict(color='blue')))

# Plot the gradient descent steps
fig.add_trace(go.Scatter(x=path, y=path_y_vals, mode='markers+lines', name='Gradient Descent Path',
                         line=dict(color='red', dash='dash'), marker=dict(color='red', size=8)))

# Highlight the starting point
fig.add_trace(go.Scatter(x=[start_x], y=[f(start_x)], mode='markers', name='Start Point',
                         marker=dict(color='green', size=12)))

# Set plot layout
fig.update_layout(title='Gradient Descent on f(x) = x^4 - 3x^3 + 2',
                  xaxis_title='x', yaxis_title='f(x)',
                  showlegend=True)

# Show the interactive plot
fig.show()
```

### Example 2: 3D

<span style="font-size: 24px;"> $y = f(x_1, x_2) = (x_1)^2 + (x_2)^2 $ <span>

```python
# Define the function and its gradient
def f(x1, x2):
    return x1**2 + x2**2

def gradient(x1, x2):
    return np.array([2*x1, 2*x2])
```

```python
# Gradient descent parameters
learning_rate = 0.1
num_iterations = 20

# Starting point
start_point = np.array([3.0, 2.0])
points = [start_point]

# Perform gradient descent
for _ in range(num_iterations):
    grad = gradient(start_point[0], start_point[1])
    start_point = start_point - learning_rate * grad
    points.append(start_point)
```

```python
# Prepare data for plotting
x1_range = np.linspace(-3, 3, 100)
x2_range = np.linspace(-3, 3, 100)
X1, X2 = np.meshgrid(x_range, y_range)
Z = f(X1, X2)

# Create the surface plot
surface = go.Surface(x=X1, y=X2, z=Z, colorscale='Viridis', opacity=0.9)
path = np.array(points)


# Create a scatter plot for the gradient descent path
scatter_path = go.Scatter3d(
    x=path[:, 0],
    y=path[:, 1],
    z=f(path[:, 0], path[:, 1]),
    mode='lines+markers',
    marker=dict(size=5, color='red'),
    name='Gradient Descent Path'
)

# Create the figure and add the surface and path
fig = go.Figure(data=[surface, scatter_path])
fig.update_layout(
    title='Gradient Descent on a 3D Surface',
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='f(X1, X2)',
        camera=dict(eye=dict(x=1.5, y=1.5, z=1.5))
    ),
    width=800,  # Set the width of the figure
    height=600  # Set the height of the figure
)

# Show the figure
fig.show()
```

### Some challenging optimization surfaces:

#### Rosenbrock Function

<span style="font-size: 24px;"> $f(x_1, x_2) = (a - x_1)^2 + b(x_2 - x_1^2)^2 $ <span>

a = 1,    b= 100

```python
# Define the Rosenbrock function
def rosenbrock(x, y, a=1, b=100):
    return (a - x)**2 + b * (y - x**2)**2

# Generate a grid of points
x = np.linspace(-2, 2, 400)
y = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Plot contour lines
ax.contour3D(X, Y, Z, levels=50, cmap='viridis', alpha=0.5)

plt.show()
```

## <center> The Anatomy of a Neural Network </center>

The most simple form of a neural network is Single-layer perceptron: 

```python
display(Image(filename = 'Perceptron.jpg', width= 1000, height= 1000))
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Generate a sample dataset
X, y = make_classification(n_samples=500, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create the neural network model with one hidden layer (e.g., 5 neurons)
mlp = MLPClassifier(hidden_layer_sizes=(5,), max_iter=1000, random_state=42)

# Fit the model to the training data
mlp.fit(X_train, y_train)

# Print the accuracy on the test set
accuracy = mlp.score(X_test, y_test)
print(f"Test Accuracy: {accuracy}")
```

```python
# Custom visualization of the neural network structure
def plot_nn_structure(model, input_dim, output_dim):
    # Get the structure of the neural network
    layer_sizes = [input_dim] + list(model.hidden_layer_sizes) + [output_dim]
    
    fig, ax = plt.subplots(figsize=(8, 6))
    v_spacing = 1.0 / (max(layer_sizes) + 1)
    h_spacing = 1.0 / (len(layer_sizes) + 1)
    
    # Positions of neurons
    neuron_positions = {}
    
    for i, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing * (layer_size - 1) / 2
        neuron_positions[i] = [(h_spacing * (i + 1), layer_top - j * v_spacing) for j in range(layer_size)]
    
    # Plot neurons
    for layer, positions in neuron_positions.items():
        for x, y in positions:
            ax.add_artist(plt.Circle((x, y), v_spacing / 10, color='b', fill=True))

    # Plot connections
    for i in range(len(layer_sizes) - 1):
        for (x1, y1) in neuron_positions[i]:
            for (x2, y2) in neuron_positions[i + 1]:
                ax.plot([x1, x2], [y1, y2], 'k-', alpha=0.4)

    ax.axis('off')
    plt.title('Neural Network Structure')
    plt.show()

# Example usage
plot_nn_structure(mlp, input_dim=2, output_dim=1)
```

This is an example of a Multilayer Perceptron (MLP) with one hidden layer. The hidden layer has 5 neurons. 

Let's estimate the total number of parameters that this model has:

- Coefficients (weights, W) = 5*2 + 5 = 15
- Intercepts (biases, b) = 5 + 1 = 6
- Total number of parameters = 15 + 6 = 21
