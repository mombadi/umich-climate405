```python
# I did not use this function because it did not provide reliable results (many nans in the output)

def cmip6_regrid(lat, lon, data, new_lat, new_lon):
    
    result = np.zeros((len(new_lat), len(new_lon), data.shape[2]))
    
    for i in range(data.shape[2]):
        
        d = data[:,:,i]
        latgrid, longrid = np.meshgrid(lat, lon, indexing="ij") #create a meshgrid using original lat and lon
        rbs = RectBivariateSpline(lat, lon, d, kx=1, ky=1) #Interpolation function
        
        n_latgrid, n_longrid = np.meshgrid(new_lat, new_lon, indexing="ij")  #create a meshgrid using new lat and lon
        interp_data = rbs.ev(n_latgrid, n_longrid) #interpolated data
        
        result[:,:,i] = interp_data
    
    return result
```
