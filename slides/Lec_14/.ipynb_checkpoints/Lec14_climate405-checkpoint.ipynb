## <center> Recurrent Neural Networks </center>

**Recurrent Neural Networks (RNNs):**  a class of neural networks designed to recognize patterns in sequences of data, such as time series, natural language, or sequences of events.

Consider a task with input (**x**) and output (**y**), then the RNN architecture has the following shape:

```python
display(Image(filename = 'RNN_architecture.png', width= 350, height= 350))
```

<span style="font-size:20px;">
In this Figure, 
    
- $x$ = Input.
- $y$ = Output.
- $h$ = Hidden state.
- $W_{xh}$ = Weights from the input to the hidden layer.
- $W_{hh}$ = Weights from $h_{t-1}$ to $h_{t}$.
- $W_{hy}$ = Weights from the hidden to the output layer.
- $b_{h}$ = bias parameter for the hidden layer.
- $b_{y}$ = bias parameter for the output layer.

</span>

The functional form of the RNN is as follows:

<span style="font-size:20px;"> $ h_t = \sigma(W_{xh} * x_t  +  W_{hh} * h_{t-1} + b_h) $  </span>

<span style="font-size:20px;"> $ y_t = \sigma(W_{hy} * h_t  + b_y) $  </span>

#### A different way to visualize an RNN (Unfold the network in the time domain):

```python
display(Image(filename = 'RNN_unfolded.png', width= 750, height= 600))
```

<span style="font-size:20px;">

In this figure:
    
- $x_{t}$ = Input at time _t_.
- $y_{t}$ = Output at time _t_.
- $h_{t} ; h_{t-1}$ = Hidden state at time _t_ and _t-1_, respectively.

</span>

### Prepare the data to be used for Machine Learning 

```python
data = pd.read_csv('jena_climate_2009_2016.csv')

# Keep only the variables and change to a numpy array
variables = data.columns[1:] #List of variables 
data = data.iloc[:,1:].values


# Scale the data relative to the mean and standard deviation of the training data 
train_data = data[1:200000,]
mean = np.mean(train_data, axis= 0)
std = np.std(train_data, axis= 0)
data = (data - mean) / std #scale the data
```

```python
data
```

### Build a custom function to generate data for training, validation and testing:

```python
def generator(data, lookback, delay, min_index, max_index=None,
              shuffle=False, batch_size=128, step=6):
    if max_index is None:
        max_index = len(data) - delay - 1
    i = min_index + lookback

    while True:
        if shuffle:
            rows = np.random.randint(min_index + lookback, max_index, size=batch_size)
        else:
            if i + batch_size >= max_index:
                i = min_index + lookback
            rows = np.arange(i, min(i + batch_size, max_index))
            i += len(rows)
        
        # Create arrays for samples and targets
        samples = np.zeros((len(rows), lookback // step, data.shape[-1]))
        targets = np.zeros(len(rows))

        for j, row in enumerate(rows):
            indices = np.arange(row - lookback, row, step)
            samples[j] = data[indices]
            targets[j] = data[row + delay, 1]  # Assuming target is in column 2

        yield samples, targets

    return samples, targets
```

#### How to use this generator function:

```python
train_gen = generator(data, lookback=1440, delay= 144, min_index= 0, max_index= 200000,
                    shuffle= True, step= 6, batch_size= 128)
```

```python
# Get the first batch of samples and targets
X_train, y_train = next(train_gen)
```

### First, Let's build a simple multi-layer preceptron model (One hidden layer with 32 nodes)

**An epoch:** refers to one complete pass through the entire training dataset. During an epoch, the model trains on all the training samples once, adjusting its weights based on the computed gradients.

```python
# Building a simple MLP model
mlp_model = Sequential()
mlp_model.add(Flatten())
mlp_model.add(Dense(units= 32))  # Hidden layer with 32 nodes. input_shape=(128, 14*240)
mlp_model.add(Dense(units = 1))  # Output layer

# Compile the model
mlp_model.compile(loss='mse', optimizer='adam')

history = mlp_model.fit(train_gen, epochs=20, steps_per_epoch= 100,
                    batch_size=32)
```

```python
# 3. Plot the loss curve
plt.plot(history.history['loss'], label='Training Loss')
#plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()
```

```python
mlp_model.summary()
```

### Now, let's build a Recurrent Neural Network (32 nodes in the hidden layer)

```python
# Building a simple RNN model
rnn_model = Sequential()
rnn_model.add(SimpleRNN(32))  # RNN with 32 units; , input_shape=(128,240, 14)
rnn_model.add(Dense(1))  # Output layer

# Compile the model
rnn_model.compile(loss='mse', optimizer='adam')

history = rnn_model.fit(train_gen, epochs=20, steps_per_epoch= 100,
                    batch_size=32)
```

```python
# Plot the loss curve
plt.plot(history.history['loss'], label='Training Loss')
#plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()
```

```python
rnn_model.summary()
```

### Let's visually compare predictions of the MLP and RNN models

Obtain predictions:

```python
test_gen = generator(data, lookback=1440, delay= 144, min_index= 300001,
                    shuffle= True, step= 6, batch_size= 128)
X_test, y_test = next(test_gen)

# MLP predictions 
mlp_pred = mlp_model.predict(X_test)
mlp_pred = mlp_pred*std[1] + mean[1]  #Rescale predictions
mlp_pred = mlp_pred.flatten()

# RNN predictions
rnn_pred = rnn_model.predict(X_test)
rnn_pred = rnn_pred*std[1] + mean[1]  #Rescale predictions 
rnn_pred = rnn_pred.flatten()

# Rescale observations 
y_test = y_test*std[1] + mean[1]
```

Plotting:

```python
# Set up the subplots with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 6))

sns.scatterplot(x=y_test, y=mlp_pred, ax=axes[0], color='blue')
axes[0].set_title('Observations vs MLP Predictions')
axes[0].set_xlabel('T_observed [C]')
axes[0].set_ylabel('T_pred [C]')
axes[0].legend(['Temperature'])

sns.scatterplot(x=y_test, y=rnn_pred, ax=axes[1], color='green')
axes[1].set_title('Observations vs RNN Predictions')
axes[1].set_xlabel('T_observed [C]')
axes[1].set_ylabel('T_pred [C]')
axes[1].legend(['Temperature'])

# Display the plots
plt.tight_layout()
plt.show()
```

## <center> In-class Excercise </center>

- Compare the predictions from the two models (MLP and RNN) with those from a Persistence model.
- Calculate the following metrics:
   - Root Mean Square Error (RMSE)
   - Nash-Sutcliffe Efficiency (NSE)
   - Pearson's correlation coefficient
   - Relative bias
- What are your observations based on these comparisons?
- How does the RMSE you calculated compare to the values shown in the loss curve? Are there any differences? If so, what might explain them?

#### Solution:

**MLP Model:**

```python
rmse = np.round(np.sqrt(np.mean( (y_test - mlp_pred)**2)), 2)
print('RMSE = ', rmse, 'C') 
print('NSE = ', 1 - (rmse/np.var(y_test))) 
print('Pearson correlation = ', np.round(np.corrcoef(y_test, mlp_pred)[0,1], 3)) 
print('Relative Bias = ', np.round((np.mean(mlp_pred - y_test)/np.mean(y_test))*100), 3, ' %') 
```

**RNN Model:**

```python
rmse = np.round(np.sqrt(np.mean( (y_test - rnn_pred)**2)), 2)
print('RMSE = ', np.round(np.sqrt(np.mean( (y_test - rnn_pred)**2)), 2), 'C') 
print('NSE = ', 1 - (rmse/np.var(y_test))) 
print('Pearson correlation = ', np.round(np.corrcoef(y_test, rnn_pred)[0,1], 3)) 
print('Relative Bias = ', np.round((np.mean(rnn_pred - y_test)/np.mean(y_test))*100), 3, ' %') 
```

Can any of the two models outperform a persistence model?

```python
persistence = y_test[:-1]
rmse = np.round(np.sqrt(np.mean( (y_test[1:] - persistence)**2)), 2)
print('RMSE = ', np.round(np.sqrt(np.mean( (y_test[1:] - persistence)**2)), 2), 'C') 
print('NSE = ', 1 - (rmse/np.var(y_test))) 
print('Pearson correlation = ', np.round(np.corrcoef(y_test[1:], persistence)[0,1], 3)) 
print('Relative Bias = ', np.round((np.mean(persistence - y_test[1:])/np.mean(y_test))*100), 3, ' %') 
```
