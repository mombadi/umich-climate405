## 1- Importing and pre-proccessing data files

### 1-1 Import Annual Max Series of Streamflow, Precip, Winter Streamflow, and Winter Precip

```python
# AMS of streamflow and precipitation
# *Note that ams_streamflow contains data for years 1981 - 2021, while ams_prcp contains data for years 1982-2022
# They also have different number of watersheds, with 3551 for streamflow and 4271 for precipitation
ams_streamflow = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Results'
                            + '/ams_streamflow.csv')
ams_prcp = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Results/'
                      + 'ams_prc.csv')

# Check
print(ams_streamflow.shape)
print(ams_prcp.shape)
```

### 1-2 Identify the sites for the case study and get the corresponding GAGES-II metadata

```python
sites = ams_streamflow.columns[1:]
sites = sites.astype(int)

metadata = pd.read_excel('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Data/GAGES_II'
                      + '/gagesII_metadata.xlsx')
metadata = metadata[metadata['STAID'].isin(sites)]
print(metadata.shape)
print(metadata.columns)
```

### 1-3 Get a list of sites based on the following characteristics: (Ref vs Non-ref), (Snowy vs. Non-snowy) and (small vs. large)

```python
ref_sites = metadata[metadata['CLASS']=='Ref']['STAID'].values
nonref_sites = metadata[metadata['CLASS']=='Non-ref']['STAID'].values
print(len(ref_sites)/metadata.shape[0] + len(nonref_sites)/metadata.shape[0])

# Snowy vs Non-snowy is defined based on whether snow fraction of annual precip is less than 5% or not
snowy = metadata[metadata['SNOW_PCT_PRECIP']>5]['STAID'].values
non_snowy = metadata[metadata['SNOW_PCT_PRECIP']<=5]['STAID'].values
print(len(snowy)/metadata.shape[0] + len(non_snowy)/metadata.shape[0])

# Catchments are either small or large based on whether the drainage basin is smaller than 1000 km2 or not
small = metadata[metadata['DRAIN_SQKM']<1000]['STAID'].values
large = metadata[metadata['DRAIN_SQKM']>=1000]['STAID'].values
print(len(small)/metadata.shape[0] + len(large)/metadata.shape[0])
```

## 2- Exploring Trends in Annual Max daily precipitaiton and streamflow

### 2-1 Extreme Daily Precipitation

```python
data = ams_prcp.iloc[:-1, 1:].to_numpy() #remove last raw (year 2022) and first column (the year column)

baseline_ams = np.nanmean(data[:20,:], axis=0) #First 20 years [1982 - 2001]
recent_ams = np.nanmean(data[20:,:], axis=0) #Second 20 years [2002 - 2021]

change = ((recent_ams - baseline_ams)/baseline_ams)*100
prcp_change = pd.DataFrame(index= np.arange(len(change)), columns= ['STAID','prcp_change'])
prcp_change['STAID'] = ams_prcp.columns[1:]
prcp_change['prcp_change'] = change

# Only keep the sites that have corresponding streamflow data (~ 3550 sites)
prcp_change['STAID'] = prcp_change['STAID'].astype(int)
prcp_change = prcp_change[prcp_change['STAID'].isin(sites)]

# Plotting 
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))
sns.histplot(ax= axes[0], data=prcp_change, x="prcp_change")

## Merge metadata with precipitation trends estimates
result = metadata.merge(prcp_change, how='outer', on='STAID')

## plot a map
from mpl_toolkits.axes_grid1 import make_axes_locatable

axes[1] = fig.add_subplot(2,1,2, projection=crs.AlbersEqualArea(central_latitude=39.5, 
                                                             central_longitude=-98.35, globe= None))
axes[1].axis('off')

axes[1].add_feature(cfeature.COASTLINE, linestyle=':', edgecolor= 'gray')
axes[1].add_feature(cfeature.STATES, linestyle=':', edgecolor= 'gray')
# Zoom in on the US by setting longitude/latitude parameters
axes[1].set_extent([-122, -72, 26, 49],crs=crs.PlateCarree())

plot = plt.scatter(x=result.LNG_GAGE, y=result.LAT_GAGE, c=result['prcp_change'], cmap='coolwarm'
            ,s= result['DRAIN_SQKM']/50,alpha=0.5,transform=crs.PlateCarree(), vmin= -20, vmax= 20) 

font = {'family':'serif','weight':'bold', 'size': 16}
plt.title('% Change in Annual Max Series of Daily Precipitaiton, n= 3550 watersheds', fontdict= font)

cbar= plt.colorbar(plot,fraction=0.046, pad=0.04, extend= 'both')
cbar.ax.tick_params(labelsize=20) 

plt.show()
```

### 2-2 Extreme Daily streamflow

```python
data = ams_streamflow.iloc[1:, 1:].to_numpy() #remove first raw (year 1981) and first column (the year column)

baseline_ams = np.nanmean(data[:20,:], axis=0) #First 20 years [1982 - 2001]
recent_ams = np.nanmean(data[20:,:], axis=0) #Second 20 years [2002 - 2021]

change = ((recent_ams - baseline_ams)/baseline_ams)*100
sf_change = pd.DataFrame(index= np.arange(len(change)), columns= ['STAID','sf_change'])
sf_change['STAID'] = ams_streamflow.columns[1:]
sf_change['sf_change'] = change

# Only keep the sites that have corresponding streamflow data (~ 3550 sites)
sf_change['STAID'] = sf_change['STAID'].astype(int)
sf_change = sf_change[sf_change['STAID'].isin(sites)]

# Plotting 
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 10))
sns.histplot(ax= axes[0], data=sf_change, x="sf_change")

## Merge metadata with precipitation trends estimates
result = metadata.merge(sf_change, how='outer', on='STAID')

## plot a map
from mpl_toolkits.axes_grid1 import make_axes_locatable

axes[1] = fig.add_subplot(2,1,2, projection=crs.AlbersEqualArea(central_latitude=39.5, 
                                                             central_longitude=-98.35, globe= None))
axes[1].axis('off')

axes[1].add_feature(cfeature.COASTLINE, linestyle=':', edgecolor= 'gray')
axes[1].add_feature(cfeature.STATES, linestyle=':', edgecolor= 'gray')
# Zoom in on the US by setting longitude/latitude parameters
axes[1].set_extent([-122, -72, 26, 49],crs=crs.PlateCarree())

plot = plt.scatter(x=result.LNG_GAGE, y=result.LAT_GAGE, c=result['sf_change'], cmap='coolwarm'
            ,s= result['DRAIN_SQKM']/50,alpha=0.5,transform=crs.PlateCarree(), vmin= -20, vmax= 20) 

font = {'family':'serif','weight':'bold', 'size': 16}
plt.title('% Change in Annual Max Series of Daily Streamflow, n= 3550 watersheds', fontdict= font)

cbar= plt.colorbar(plot,fraction=0.046, pad=0.04, extend= 'both')
cbar.ax.tick_params(labelsize=20) 

plt.show()
```

### plotting function

```python
def plotting_trend(df, title):
    
    years = np.arange(1981, 2022)
    y = np.nanmedian(df, axis= 1)
    y_25 = np.nanpercentile(df, 25, axis = 1)
    y_75 = np.nanpercentile(df, 75, axis = 1)
    
    plt.plot(years, y, color= 'black')
    plt.fill_between(years,y_25, y_75, alpha= 0.2, color= 'gray')
    
    elnino_years = np.array([1982, 1983, 1997, 1998, 2015, 2016])
    lanina_years = np.array([1988, 1989, 1999, 2000, 2007, 2008, 2010, 2011])
    
    plt.plot(elnino_years, y[elnino_years - 1981], 'o', color= 'red')
    plt.plot(lanina_years, y[lanina_years - 1981], 'o', color= 'blue')
    
    plt.title(title + ';  n=' + str(df.shape[1]))
    
    plt.tight_layout()
    plt.savefig('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Figures/'
               + title + '.jpeg', dpi= 600)

    
    plt.show()
```

## Trend analysis (Standardized streamflow, z= (x - u)/sigma)

### Standardization data

```python
# Compute mean and std of daily streamflow
streamflow = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/' +
                         'Data/USGS_streamflow/streamflow_v1.csv')
streamflow = streamflow.drop('datetime', axis=1)
mean_streamflow = np.nanmean(streamflow, axis=0)
std_streamflow = np.nanstd(streamflow, axis=0)

# Annual Max Streamflow
ams_streamflow = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Results'
                            + '/ams_winter_streamflow.csv')
ams_streamflow = ams_streamflow.drop(columns=ams_streamflow.columns[0], axis=1)

# Standardize annual max streamflow
ams_streamflow = (ams_streamflow - mean_streamflow) / std_streamflow
```

```python
# Ref vs Non-ref
ref_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(ref_sites)]
nonref_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(nonref_sites)]

# Snow-dominated vs Non-snowy
snowy_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(snowy)]
nonsnowy_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(non_snowy)]

# Large vs Small
small_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(small)]
large_ams = ams_streamflow.iloc[:,ams_streamflow.columns.astype(int).isin(large)]
```

### plotting

```python
#fig, axs = plt.subplots(2, 2)

plotting_trend(ref_ams, 'Annual AMS, Ref sites')
plotting_trend(nonref_ams, 'Annual AMS, Non-ref sites')

plotting_trend(snowy_ams, 'Annual AMS, Snowy sites')
plotting_trend(nonsnowy_ams, 'Annual AMS, Non-snowy sites')

plotting_trend(small_ams, 'Annual AMS, Small sites')
plotting_trend(large_ams, 'Annual AMS, Large sites')
```

### plotting - Winter streamflow

```python
#fig, axs = plt.subplots(2, 2)

plotting_trend(ref_ams, 'Winter AMS, Ref sites')
plotting_trend(nonref_ams, 'Winter AMS, Non-ref sites')

plotting_trend(snowy_ams, 'Winter AMS, Snowy sites')
plotting_trend(nonsnowy_ams, 'Winter AMS, Non-snowy sites')

plotting_trend(small_ams, 'Winter AMS, Small sites')
plotting_trend(large_ams, 'Winter AMS, Large sites')
```
