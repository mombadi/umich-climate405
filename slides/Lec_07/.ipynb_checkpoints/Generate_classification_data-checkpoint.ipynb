```python
# Import libraries
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from scipy.integrate import odeint
from IPython.display import Image
from statistics import mode
from scipy.stats import pearsonr, spearmanr
from sklearn.feature_selection import mutual_info_regression
import statsmodels.api as sm
import pingouin as pg

# Suppress warnings
warnings.filterwarnings('ignore')

# Set number of decimals for np print options
np.set_printoptions(precision=3)

# Set the current working directory
os.chdir(sys.path[0])
```

```python
from sklearn.datasets import make_classification

#plt.title("Two informative features, two clusters per class", fontsize="small")
X, y = make_classification(n_features=2, n_redundant=0, n_informative=2)
df = pd.DataFrame(np.column_stack((X,y)), columns= list(['x1', 'x2', 'y']))
```

## Plot Full Dataset

```python
color_map = {0: 'blue', 1: 'red', 2: 'white'}
df['color'] = df['y'].map(color_map)
plt.scatter(df['x1'], df['x2'], marker="o", c=df['color'], s=45, edgecolor="k")
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Full Dataset')
plt.tight_layout()
plt.savefig('Full Dataset_Example.jpeg', dpi= 600)
plt.show()
```

### Plot Supervised Learning Example

```python
# Randomly unlabel some points 
df['y'].iloc[:30] = 2
```

```python
color_map = {0: 'blue', 1: 'red', 2: 'white'}
df['color'] = df['y'].map(color_map)

plt.scatter(df['x1'], df['x2'], marker="o", c=df['color'], s=45, edgecolor="k")
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Supervised Learning')
plt.tight_layout()
plt.savefig('Supervised Learning_Example.jpeg', dpi= 600)
plt.show()
```

### Plot Unsupervised Learning Example

```python
color_map = {0: 'blue', 1: 'red', 2: 'white'}
df['color'] = df['y'].map(color_map)

plt.scatter(df['x1'], df['x2'], marker="o", c='white', s=45, edgecolor="k")
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Supervised Learning')
plt.tight_layout()
plt.savefig('Unsupervised Learning_Example.jpeg', dpi= 600)
plt.show()
```
