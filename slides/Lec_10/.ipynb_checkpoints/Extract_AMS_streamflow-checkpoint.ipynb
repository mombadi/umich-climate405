## Extract AMS of Winter Streamflow

```python
## import dataset 
streamflow = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/'
                    + 'Data/USGS_streamflow/streamflow_v1.csv')
streamflow['datetime'] = streamflow['datetime'].astype(str)
streamflow['year'] = np.nan
streamflow['month'] = np.nan
streamflow['hy_year'] = np.nan

# Extract year, month and hyrdologic year
for i in range(streamflow.shape[0]):
    streamflow['year'][i] = int(streamflow['datetime'][i][:4])
    streamflow['month'][i] = int(streamflow['datetime'][i][5:7])
    if streamflow['month'][i] >= 10:
        streamflow['hy_year'][i] = streamflow['year'][i] + 1
    else:
        streamflow['hy_year'][i] = streamflow['year'][i]

# Extract AMS
ams_winter_streamflow = pd.DataFrame(index = np.arange(1981, 2022), columns= streamflow.columns[1:-3])
years = np.arange(1981, 2022)
for i in range(len(years)):
    d = streamflow[(streamflow['hy_year'] == years[i]) & (streamflow['month'].isin([10, 11, 12, 1, 2]))]
    d = d.drop(columns = ['datetime', 'year', 'month', 'hy_year'])
    ams_winter_streamflow.iloc[i,:] = np.nanmax(d, axis= 0)
```

## Extract AMS of daily precipitaiton

```python
mat = scipy.io.loadmat('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Data/PRISM/' 
              + 'prism_watershed_av_prcp.mat')
daily_prc = mat['watershed_prcp']

# add site ids
sites = pd.read_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding/Data/PRISM/'
                    + 'sites.csv')

# change to a dataframe
daily_prc = pd.DataFrame(daily_prc, columns= sites.columns)

# Add datetime, year, month and hy-year
t = np.arange(datetime(1981,1,1), datetime(2023,12,31), timedelta(days=1)).astype(datetime)
daily_prc['datetime'] = t
daily_prc['year'] = pd.DatetimeIndex(daily_prc['datetime']).year
daily_prc['month'] = pd.DatetimeIndex(daily_prc['datetime']).month
daily_prc['hy_year'] = np.nan

for i in range(daily_prc.shape[0]):
    if daily_prc['month'][i] >= 10:
        daily_prc['hy_year'][i] = daily_prc['year'][i] + 1
    else:
        daily_prc['hy_year'][i] = daily_prc['year'][i]
```

```python
# Extract AMS
ams_prc = pd.DataFrame(index = np.arange(1982, 2023), columns= daily_prc.columns[0:-4])
years = np.arange(1982, 2023)
for i in range(len(years)):
    d = daily_prc[daily_prc['hy_year'] == years[i]]
    d = d.drop(columns = ['datetime', 'year', 'month', 'hy_year'])
    ams_prc.iloc[i,:] = np.nanmax(d, axis= 0)
```

```python
# save ams_prc
ams_prc.to_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding' + 
                     '/Results/ams_prc.csv')
```
