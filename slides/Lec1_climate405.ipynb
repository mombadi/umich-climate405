# <center>Stochastic vs Deterministic Systems</center>

### Stochastic system: 
A (temporal) sequence of **<i>random variables<i>**, with each random variable taking values from the state space.

Examples:
- Wiener Process.
- Poisson Process.
- Random Walk.
- Markov Chain


```python
plt.plot(np.arange(100), np.random.normal(0,1,100), '-o')
```

### Deterministic system:
A system that its future states depend on initial conditions, with **<i>no randomness<i>** involved.

Examples:
- Chaotic systems (e.g., Lorenz system (aka the Butterfly attractor)

### Let's model this in a function:

```python
def markov_chain(initial_state, n, P):
    
    #Simulates a Markov chain.

    #Parameters:
    #initial_state (str): The initial state of the chain ('Hot' or 'Cold').
    #n (int): The number of states to simulate.
    #P (numpy.ndarray): The transition probability matrix.
    
    #Returns:
    #list: A sequence of states over n transitions.
    #----------------------------------------------------
    
    # Define state names and map them to indices
    states = ['Hot', 'Cold']
    state_index = {state: idx for idx, state in enumerate(states)}
    
    # Map the initial state to its index
    current_state_idx = state_index[initial_state]
    
    # Initialize the sequence with the initial state
    sequence = [states[current_state_idx]]
    
    for _ in range(n - 1):
        # Get the transition probabilities for the current state
        transition_probs = P[current_state_idx]
        
        # Choose the next state based on the transition probabilities
        next_state_idx = np.random.choice([0, 1], p=transition_probs)
        
        # Append the next state to the sequence
        sequence.append(states[next_state_idx])
        
        # Update the current state index
        current_state_idx = next_state_idx

    return sequence
```

### Now, run the function with the values of transition probabilities above:

```python
sim_1 = markov_chain('Hot', 20, np.array([
    [0.25, 0.75],
    [0.7, 0.3]]))

sim_2 = markov_chain('Hot', 20, np.array([
    [0.25, 0.75],
    [0.7, 0.3]]))

df = pd.DataFrame({'sim_1': sim_1,
        'sim_2': sim_2})
```

## <center>Deterministic System: Lorenz Attractor</center>

```python
display(Image(filename = 'Ed Lorenz+PDEs.png', width= 750, height= 250))
```

### Let's model this in a function:

```python
# Define the Lorenz system of differential equations
def lorenz(state, t, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return dxdt, dydt, dzdt
```

### Run the function:

```python
# Parameters for the Lorenz system
sigma = 10
rho = 28
beta = 8/3

# Initial state
state0 = [0, 1, 1]

# Time vector
t = np.linspace(0, 20, 20000)

# Solve ODE
states = odeint(lorenz, state0, t, args=(sigma, rho, beta))
```

### Plotting:

```python
# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(states[:, 0], states[:, 1], states[:, 2], lw=0.5)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Lorenz Attractor')

plt.show()
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)
ax.plot(states[:, 1], states[:, 2], lw=0.5)

ax.set_xlabel('Y')
ax.set_ylabel('Z')
ax.set_title('Lorenz Attractor')

plt.show()
```

# <center>Random Variables</center>

### Let's consider the following time series as a random variable:
This is average daily wind speed (m/s) data for July, 2024 in Central Park, New York


```python
data = pd.read_csv('wind speed_Central Park_NY.csv')

print(data['AWND'].values)

plt.plot(np.arange(data.shape[0]), data['AWND'].values, '-o')
plt.show()

plt.hist(data['AWND'])
plt.show()
```

### 1st moment: Mean

```python
plt.hist(data['AWND'])
plt.vlines(x= np.mean(data['AWND']), ymin= 0, ymax= 8, color= 'red')
plt.text(1.5, 8, 'Mean = ' + str(round(np.mean(data['AWND']), 1)), color= 'red')

plt.show()
```

### Other measures of central tendency

#### Median

```python
plt.hist(data['AWND'])
plt.vlines(x= np.median(data['AWND']), ymin= 0, ymax= 8, color= 'red')
plt.text(1.5, 8, 'Median = ' + str(round(np.median(data['AWND']), 1)), color= 'red')

plt.show()
```

#### Mode

```python
plt.hist(data['AWND'])
plt.vlines(x= mode(data['AWND']), ymin= 0, ymax= 8, color= 'red')
plt.text(1.5, 8, 'Mode = ' + str(round(mode(data['AWND']), 1)), color= 'red')

plt.show()
```

### 2nd Moment: Variance

```python
print('Variance = ' + str(np.var(data['AWND'])))
print('Standard Deviation= ' + str(np.sqrt(np.var(data['AWND']))))
```

### 3rd Moment: Skewness

```python
print('Skewness = ' + str(skew(data['AWND'])))
```

### 4th Moment: Kurtosis
Characterizes the heaviness of the distribution's tail

```python
print('Kurtosis = ' + str(kurtosis(data['AWND'])))
```
