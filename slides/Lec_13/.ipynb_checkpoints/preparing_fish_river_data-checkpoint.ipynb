```python
streamflow = pd.read_csv('/Users/ombadi/Desktop/Mo Files/Research/USGS_streamflow/streamflow_v1.csv')
sites = streamflow.columns[1:]
```

```python
from datetime import datetime, timedelta
```

```python
# Fish River near Fort Kent, Maine
# Area = 2252.7 km^2
runoff_data = streamflow[['datetime', '01013500']]

# Import rainfall data, and extrat it for the same watershed 
from scipy.io import loadmat
precip = loadmat('/Users/ombadi/Desktop/Mo Files/Research/PRISM/prism_watershed_av_prcp.mat')['watershed_prcp']

# add site ids
sites = pd.read_csv('/Users/ombadi/Desktop/Mo Files/Research/PRISM/sites.csv')

# change to a dataframe
daily_prc = pd.DataFrame(precip, columns= sites.columns)

# Add datetime, year, month and hy-year
t = np.arange(datetime(1981,1,1), datetime(2023,12,31), timedelta(days=1)).astype(datetime)
daily_prc['datetime'] = t
daily_prc['year'] = pd.DatetimeIndex(daily_prc['datetime']).year
daily_prc['month'] = pd.DatetimeIndex(daily_prc['datetime']).month

# Rainfall data
rainfall_data = daily_prc[['datetime', 'year','01013500']]
```

```python
#runoff_data['datetime'] = pd.to_datetime(runoff_data['datetime'])
#runoff_data.dtypes

fish_river_data = rainfall_data.merge(runoff_data, on='datetime', how='inner')
fish_river_data = fish_river_data.rename(columns={"01013500_x": "rainfall", "01013500_y": "runoff"})
fish_river_data = fish_river_data[(fish_river_data['year'] >= 2016) & (fish_river_data['year'] <= 2021)]
fish_river_data.index = np.arange(fish_river_data.shape[0])

fish_river_data.to_csv('/Users/ombadi/Desktop/Mo Files/Teaching/CLIMATE 405/Slides/Lec 10/fish_river.csv', index= False)

fish_river_data.head()
```
