## Import Libraries

```python
# Basic libraries
import os
import glob
import numpy as np
import pandas as pd

# Plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Other
import netCDF4 as nc
import calendar
import numpy.ma as ma
from scipy.interpolate import RectBivariateSpline
from scipy.io import savemat
 
%matplotlib inline
pd.options.mode.chained_assignment = None  # default='warn'

import cartopy
import cartopy.crs as crs
import cartopy.feature as cfeature
```

```python
## Import all files for USGS Stations and calculate the number of years with acceptable sample size

files = glob.glob('/Users/ombadi/Desktop/MO Files/Research/LBL_Research/Human impact on modifying drought severity'
                  + '/Data/USGS_streamflow_processed_daily/All/*.csv')
#len(files) = 8001 files

df0 = pd.read_csv(files[0])
data = pd.DataFrame(index = df0['datetime'], columns= np.arange(len(files)))
sites = []
n = []
col_index = 0

for f in range(len(files)):
    
    df = pd.read_csv(files[f])
    
    if (df['year'][0] > 1980) | (df['year'][df.shape[0]-1] < 2020):
        continue
    
    sites.append(os.path.basename(files[f])[:-4]) #save usgs site id
    data.iloc[:df.shape[0],col_index] = df['Q'].values #save daily streamflow from each site
    col_index = col_index + 1 #update col_index
    
    #Calculate the number of years with acceptable number of observations (n_site)
    n_site = 0 
    for i in np.arange(1980, 2022):
        
        x = df[df['hy_year']==i]
        n_site = n_site + (np.nansum(x['missing']) < 365*0.1)
        
    n.append(n_site)
    
    print(f)
```

```python
# Number of sites that have data between 1980 and 2019 (40 years period)
#print(col_index), 3934

# Save data for those 3934 sites 
streamflow_v0 = pd.DataFrame(index = data.index, columns = sites)
streamflow_v0[:] = data.iloc[:, :col_index]
streamflow_v0
streamflow_v0.to_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding' + 
                     '/Data/USGS_streamflow/streamflow_v0.csv')
```

```python
# Remove sites that have less than 30 acceptable years
list_sites = np.array(sites)
#len(list_sites[np.array(n) < 30]); 384 sites

columns_drop = list_sites[np.array(n) < 30]
streamflow_v1 = streamflow_v0.copy()
streamflow_v1 = streamflow_v1.drop(columns= columns_drop)
streamflow_v1.to_csv('/Users/ombadi/Desktop/MO Files/Research/Antecedent conditions and flooding' + 
                     '/Data/USGS_streamflow/streamflow_v1.csv')
```
